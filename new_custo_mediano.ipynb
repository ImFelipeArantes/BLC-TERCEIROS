```python
#### LAST MILE
# Pré-processa os dados de valores_ethernet em um dicionário para acesso rápido
valores_dict = {}
for velocidade in [10, 50, 100, 200, 300, 400, 500, 1000]:
    subset = valores_ethernet[valores_ethernet['VEL'] == velocidade]
    valores_dict[velocidade] = subset.groupby('SIGLA_MUNICIPIO').agg({
        'CUSTO_MENSAL': 'median',
        'TAXA_INSTALACAO': 'median'
    }).rename(columns={
        'CUSTO_MENSAL': 'CUSTO_MENSAL_MEDIAN',
        'TAXA_INSTALACAO': 'TAXA_INSTALACAO_MEDIAN'
    })

# Para cada velocidade, calcula os valores e adiciona as colunas no df
for velocidade in [10, 50, 100, 200, 300, 400, 500, 1000]:
    print(f"Processando {velocidade} Mbps")
    dados_vel = valores_dict[velocidade]

    # Faz merge entre o df e os valores medianos da velocidade atual
    df_merged = df.merge(
        dados_vel,
        how='left',
        left_on='CNL',
        right_index=True
    )

    # Calcula os valores
    df[f'CM_LM_{velocidade}M_INSTALACAO'] = df_merged['TAXA_INSTALACAO_MEDIAN'].fillna(df_merged['TAXA_INSTALACAO_MEDIAN'].max())
    df[f'CM_LM_{velocidade}M_MENSAL'] = df_merged['CUSTO_MENSAL_MEDIAN'].fillna(df_merged['CUSTO_MENSAL_MEDIAN'].max())
    df[f'CM_LM_{velocidade}M'] = (
        df[f'CM_LM_{velocidade}M_MENSAL'] +
        df[f'CM_LM_{velocidade}M_INSTALACAO'] / 24
    )


### BANDA LARGA
# Pré-processa os dados de valores_ethernet em um dicionário para acesso rápido
valores_dict = {}
for velocidade in [150, 300, 600, 1000]:
    subset = valores_internet[valores_internet['VEL'] == velocidade]
    valores_dict[velocidade] = subset.groupby('SIGLA_MUNICIPIO').agg({
        'CUSTO_MENSAL': 'median',
        'TAXA_INSTALACAO': 'median'
    }).rename(columns={
        'CUSTO_MENSAL': 'CUSTO_MENSAL_MEDIAN',
        'TAXA_INSTALACAO': 'TAXA_INSTALACAO_MEDIAN'
    })

# Para cada velocidade, calcula os valores e adiciona as colunas no df
for velocidade in [150, 300, 600, 1000]:
    print(f"Processando {velocidade} Mbps")
    dados_vel = valores_dict[velocidade]

    # Faz merge entre o df e os valores medianos da velocidade atual
    df_merged = df.merge(
        dados_vel,
        how='left',
        left_on='CNL',
        right_index=True
    )

    # Calcula os valores
    df[f'CM_BL_{velocidade}M_INSTALACAO'] = df_merged['TAXA_INSTALACAO_MEDIAN'].fillna(df_merged['TAXA_INSTALACAO_MEDIAN'].max())
    df[f'CM_BL_{velocidade}M_MENSAL'] = df_merged['CUSTO_MENSAL_MEDIAN'].fillna(df_merged['CUSTO_MENSAL_MEDIAN'].max())
    df[f'CM_BL_{velocidade}M'] = (
        df[f'CM_BL_{velocidade}M_MENSAL'] +
        df[f'CM_BL_{velocidade}M_INSTALACAO'] / 24
    )
```
